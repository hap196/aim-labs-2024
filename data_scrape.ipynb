```python
from selenium.webdriver.common.by import By
from selenium import webdriver
import json
from bs4 import BeautifulSoup
import requests
import re
import time
from fake_useragent import UserAgent
```

```python
all_links = []
for num in range(1, 2):
    driver = webdriver.Chrome()
    driver.get('https://stockx.com/sneakers?page=' + str(num))
    links = driver.find_elements(By.TAG_NAME, "a")
    for link in links:
        all_links.append(link.get_attribute('href'))
    # driver.quit()
```

```python
urls = []

for i in range(1, 6):
    with open(f'page_sources/{i}.html', 'r') as f:
        contents = f.read()

    soup = BeautifulSoup(contents, 'html.parser')
    elements = soup.select('[data-testid="productTile-ProductSwitcherLink"]')

    for element in elements:
        urls.append(element.get('href'))
```

```python
def write_links(urls, output_file):
    with open(output_file, 'w') as f:
        for url in urls:
            # Write the link to the file, prefixed with 'stockx.com/'
            f.write('https://stockx.com' + url + '\n')
```
